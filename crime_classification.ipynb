```python
import pandas as pd
import numpy as np
import tools as t
import time 

np.random.seed(seed=99)
%pylab inline
```

```python
train_data = pd.read_csv("train.csv")
train_data
```

## Traing only on Latitude and longitude

```python
X = np.array(train_data[["X", "Y"]])
Y = np.array(train_data["Category"])

xtrain, ytrain, xtest,ytest = t.split_data(X,Y)

print xtrain.shape, ytrain.shape
print xtest.shape, ytest.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
%pdb OFF
rf = RandomForestClassifier()
rf.fit(xtrain, ytrain)
```

```python
Yp=rf.predict(xtest)
print ytest.shape, len(Yp)
t.print_confusion_matrix(Yp,ytest)
print sum(ytest == np.array(Yp)) / float(ytest.shape[0])
```

## Traing on Descript , DayOfWeek , PdDistrict , Resolution , X , Y 

```python
X = np.array(train_data[["Descript","DayOfWeek","PdDistrict","Resolution","X","Y"]])
Y = np.array(train_data["Category"])
```

```python
from sklearn import preprocessing
enc = preprocessing.LabelEncoder()

X[:,0] = enc.fit_transform(X[:,0])
X[:,1] = enc.fit_transform(X[:,1])
X[:,2] = enc.fit_transform(X[:,2])
X[:,3] = enc.fit_transform(X[:,3])

xtrain, ytrain, xtest,ytest = t.split_data(X,Y)

print xtrain.shape, ytrain.shape
print xtest.shape, ytest.shape
```

```python
%pdb OFF
rf = RandomForestClassifier()
rf.fit(xtrain, ytrain)
```

```python
Yp=rf.predict(xtest)
print ytest.shape, len(Yp)
print sum(ytest == np.array(Yp)) / float(ytest.shape[0])
```

```python
print sum(ytest == np.array(Yp)) / float(ytest.shape[0])
t.print_confusion_matrix(Yp,ytest)
```
